```python


def combiner(pathos, type='csv', axis_align=0):
    listo = []
    iterrer = pathlib.Path(pathos)
    fillos = list(iterrer.rglob(f"*.{type}"))


    for thingo in fillos:
        inter = pd.read_csv(thingo)

        if axis_align == 1: 
            cols = inter.columns.tolist()
            inter.set_index(cols[0], inplace=True)
        
        listo.append(inter)

    if axis_align == 0:
        cat = pd.concat(listo)
    else:
        cat = pd.concat(listo, axis=1)
        cat.reset_index(inplace=True)

    pp(cat)


combiner('inter/underlying_causes', 'csv', 1)
```
